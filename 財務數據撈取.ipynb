```python
#會計科目欄位查詢
accounting = tejapi.get('TWN/AIACC',
                       id = 'AIM1A',  #這邊要注意一下，因為要挑選簡表欄位
                       paginate = True,
                       chinese_column_name = True)
```

```python
accounting
```

```python
#查詢對應科目的方式(用中文全稱or英文全稱)，可順便留意單位
#折舊
depreciation = accounting[accounting['中文全稱'].str.contains('折舊')]
depreciation
```

```python
#找出資產負債表所有科目
#balance_sheet 
balance_sheet = accounting[accounting['中文表次'].str.contains('資產負債')]
balance_sheet
```

```python
BS_list = balance_sheet['會計科目'].tolist()
BS_list
```

```python
finance = pd.DataFrame()
for firm in panel:
    finance = finance.append(tejapi.get('TWN/AIM1A',
                    coid = firm,
                    paginate = True,
                    opts = {'pivot':True, 'columns': ['coid','mdate'] + BS_list},
                    chinese_column_name = True))
finance
```

```python
finance = finance.reset_index(drop = True)
finance
```

```python
#備份
fin = finance.copy()
fin
```
